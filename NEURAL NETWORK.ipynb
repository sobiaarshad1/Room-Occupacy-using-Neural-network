```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from numpy import random

#read  data using numpy
df = pd.read_csv('Roomoccupy.csv', delimiter= ',')
org_df = df.copy()
#noramalization of data

df=(df-df.min())/(df.max()-df.min())
# Display
df

df=df.iloc[0:,[0,1,2,3,4,5]]
df=pd.DataFrame(df)

# adding any 5 features in x_values array
x_values=df.iloc[0:,0:5]
x_values = np.array(x_values)

# adding the predictive feature in y_values array
y_values =df.iloc[0:,5]
y_values= np.array(y_values)


#for 2 hidden units 
columns=5
v1=np.zeros(columns)
v2=np.zeros(columns)
w=np.zeros(2)
#eta=0.01
eta=0.1
#eta=0.05
#eta=0.005
k=2
iteration=50

def sigmoidFunc1(x_values):
    sum1=0
    for j in range(5):   
        sum1 += x_values[j]
    sigmoid=1/(1 + np.exp(-sum1))
    return sigmoid

def sigmoidFunc(x_values):
    sigmoid=1/(1 + np.exp(-x_values))
    return sigmoid

# for 2 hidden units
def cal_hidden_layer(x_values,y_values,v1,v2,w,k,eta,iteration):
    v_array=[]
    array1=[]
    v_array.insert(0,v1)
    v_array.insert(1,v2)
    h_array=[]
    dw=np.zeros(2)
    dv=np.zeros(2)
    for m in range(iteration):
        total_loss=0
        for i in range(2664):
            if k==2:
                fvw=0
                val1=np.dot(v_array[0],x_values[i])
                val2=np.dot(v_array[1],x_values[i])
                h_array.clear()
                h1=sigmoidFunc(val1)
                h2=sigmoidFunc(val2)
                h_array.insert(0,h1)
                h_array.insert(1,h2)
                h=np.array(h_array)
                for j in range(k):
                    z=np.dot(v_array[j],x_values[i])
                    b=sigmoidFunc(z) 
                    fvw += w[j]*b
                    fw=sigmoidFunc(fvw)
                    dw[j] = 2*(fw-y_values[i])*fw*(1-fw)*sigmoidFunc(np.dot(v_array[j],x_values[j]))          
                    w[j]=w[j]-eta*dw[j]   
                    dv[j] = 2*(fw - y_values[i])*fw*(1-fw)*b*(1-b)*w[j]
                    v_array[j]=v_array[j]-eta*dv[j]
                w_h=np.dot(w,h)
                f= sigmoidFunc(w_h)
                total_loss += (f-y_values[i])**2
        train_loss= (1/2664)*total_loss
        array1.insert(m,train_loss)
    array1=np.array(array1)
    return array1

h=cal_hidden_layer(x_values,y_values,v1,v2,w,k,eta,iteration)
y=h
plt.rcParams["figure.figsize"] = [7.50, 5.50]
plt.rcParams["figure.autolayout"] = True

it_arr = []
for i in range(0,iteration):
    t = i
    it_arr.append(t)
    
iteration_array=np.array(it_arr)
x=iteration_array
plt.plot(x, y, color="red")


                                 # for 3 hidden units
    
columns=5
#eta=0.01
eta=0.1
#eta=0.05
#eta=0.005
k=3
iteration=80

v1=np.zeros(columns)
v2=np.zeros(columns)
v3=np.zeros(columns)
w=np.zeros(3)

def cal_hidden_layer3(x_values,y_values,v1,v2,v3,w,k,eta,iteration):
    v_array=[]
    array1=[]
    v_array.insert(0,v1)
    v_array.insert(1,v2)
    v_array.insert(2,v3)
    h_array=[]
    dw=np.zeros(3)
    dv=np.zeros(3)
    for m in range(iteration):
        total_loss=0
        for i in range(2664):
            if k==3:
                fvw=0
                val1=np.dot(v_array[0],x_values[i])
                val2=np.dot(v_array[1],x_values[i])
                h_array.clear()
                h1=sigmoidFunc(val1)
                h2=sigmoidFunc(val2)
                h_array.insert(0,h1)
                h_array.insert(1,h2)
                h_array.insert(2,h2)
                h=np.array(h_array)
                for j in range(k):
                    z=np.dot(v_array[j],x_values[i])
                    b=sigmoidFunc(z) 
                    fvw += w[j]*b
                    fw=sigmoidFunc(fvw)
                    dw[j] = 2*(fw-y_values[i])*fw*(1-fw)*sigmoidFunc(np.dot(v_array[j],x_values[j]))          
                    w[j]=w[j]-eta*dw[j]   
                    dv[j] = 2*(fw - y_values[i])*fw*(1-fw)*b*(1-b)*w[j]
                    v_array[j]=v_array[j]-eta*dv[j]
                w_h=np.dot(w,h)
                f= sigmoidFunc(w_h)
                total_loss += (f-y_values[i])**2
        train_loss= (1/2664)*total_loss
        array1.insert(m,train_loss)
    array1=np.array(array1)
    return array1


h=cal_hidden_layer3(x_values,y_values,v1,v2,v3,w,k,eta,iteration)
y=h
it_arr = []
for i in range(0,iteration):
    t = i
    it_arr.append(t)
    
iteration_array=np.array(it_arr)
x=iteration_array
plt.plot(x, y, color="yellow")


                                      # for 4 hidden units

columns=5
#eta=0.05
#eta=0.01
eta=0.1
#eta=0.005
k=4
it=100


v1=np.zeros(columns)
v2=np.zeros(columns)
v3=np.zeros(columns)
v4=np.zeros(columns)
w=np.zeros(4)
def cal_hidden_layer4(x_values,y_values,v1,v2,v3,v4,w,k,eta,iteration):
    v_array=[]
    array1=[]
    v_array.insert(0,v1)
    v_array.insert(1,v2)
    v_array.insert(2,v3)
    v_array.insert(3,v3)
    h_array=[]
    dw=np.zeros(4)
    dv=np.zeros(4)
    for m in range(iteration):
        total_loss=0
        for i in range(2664):
            if k==4:
                fvw=0
                val1=np.dot(v_array[0],x_values[i])
                val2=np.dot(v_array[1],x_values[i])
                h_array.clear()
                h1=sigmoidFunc(val1)
                h2=sigmoidFunc(val2)
                h_array.insert(0,h1)
                h_array.insert(1,h2)
                h_array.insert(2,h2)
                h_array.insert(3,h2)
                h=np.array(h_array)
                for j in range(k):
                    z=np.dot(v_array[j],x_values[i])
                    b=sigmoidFunc(z) 
                    fvw += w[j]*b
                    fw=sigmoidFunc(fvw)
                    dw[j] = 2*(fw-y_values[i])*fw*(1-fw)*sigmoidFunc(np.dot(v_array[j],x_values[j]))          
                    w[j]=w[j]-eta*dw[j]   
                    dv[j] = 2*(fw - y_values[i])*fw*(1-fw)*b*(1-b)*w[j]
                    v_array[j]=v_array[j]-eta*dv[j]
                w_h=np.dot(w,h)
                f= sigmoidFunc(w_h)
                total_loss += (f-y_values[i])**2
        train_loss= (1/2664)*total_loss
        array1.insert(m,train_loss)
    array1=np.array(array1)
    return array1


h=cal_hidden_layer4(x_values,y_values,v1,v2,v3,v4,w,k,eta,iteration)
y=h
it_arr = []
for i in range(0,iteration):
    t = i
    it_arr.append(t)
    
iteration_array=np.array(it_arr)
x=iteration_array

plt.plot(x, y, color="green")

                                         #for 5 hidden units
    
k=5
#eta=0.01
eta=0.1
#eta=0.05
#eta=0.005
k=5
iteration=100
columns=5


v1=np.zeros(columns)
v2=np.zeros(columns)
v3=np.zeros(columns)
v4=np.zeros(columns)
v5=np.zeros(columns)
w=np.zeros(columns)
def cal_hidden_layer5(x_values,y_values,v1,v2,v3,v4,v5,w,k,eta,iteration):
    v_array=[]
    array1=[]
    v_array.insert(0,v1)
    v_array.insert(1,v2)
    v_array.insert(2,v3)
    v_array.insert(3,v3)
    v_array.insert(4,v3)
    h_array=[]
    dw=np.zeros(5)
    dv=np.zeros(5)
    for m in range(iteration):
        total_loss=0
        for i in range(2664):
            if k==5:
                fvw=0
                val1=np.dot(v_array[0],x_values[i])
                val2=np.dot(v_array[1],x_values[i])
                h_array.clear()
                h1=sigmoidFunc(val1)
                h2=sigmoidFunc(val2)
                h_array.insert(0,h1)
                h_array.insert(1,h2)
                h_array.insert(2,h2)
                h_array.insert(3,h2)
                h_array.insert(4,h2)
                h=np.array(h_array)
                for j in range(k):
                    z=np.dot(v_array[j],x_values[i])
                    b=sigmoidFunc(z) 
                    fvw += w[j]*b
                    fw=sigmoidFunc(fvw)
                    dw[j] = 2*(fw-y_values[i])*fw*(1-fw)*sigmoidFunc(np.dot(v_array[j],x_values[j]))          
                    w[j]=w[j]-eta*dw[j]   
                    dv[j] = 2*(fw - y_values[i])*fw*(1-fw)*b*(1-b)*w[j]
                    v_array[j]=v_array[j]-eta*dv[j]
                w_h=np.dot(w,h)
                f= sigmoidFunc(w_h)
                total_loss += (f-y_values[i])**2
        train_loss= (1/2664)*total_loss
        array1.insert(m,train_loss)
    array1=np.array(array1)
    return array1


h=cal_hidden_layer5(x_values,y_values,v1,v2,v3,v4,v5,w,k,eta,iteration)
y=h
it_arr = []
for i in range(0,iteration):
    t = i
    it_arr.append(t)
    
iteration_array=np.array(it_arr)
x=iteration_array
plt.title("Neural Network")
plt.xlabel("Iterations")
plt.ylabel("train loss")
plt.plot(x, y, color="blue")
```
